```python
video_dir = r'C:\Users\USER\OneDrive - Stellenbosch University\Masters\Thesis\virtual_envs\coldb_files\four_models\Cascade R-CNN_V3\val'
frame_error = []
order_list = [100,200,300,400,500,600]
order_dict = {}
for order in order_list:
    print(str(order) + "...")

    # theta = np.polyfit(count_c,thresh_c , order)
    # p = np.poly1d(theta)
    #opt_thresh = p(count)
    all_vid_errors = []
    error_dict = {}
    for vid in os.listdir(video_dir):
        
        #if vid == 'DJI_0829-552':#!= 'Altitude_videos' and vid != 'dicts' and vid != 'Circle_vid' and vid != 'val':
            
        print(vid)
        vid_path = os.path.join(video_dir, vid)

        #Starting threshold for each video
        threshold = 0.99

        # Get ground truth of saving later
        ground_truth = float(vid.split("-")[1])
        #Define list of counts for each video
        counts = []
        # Define list of errors for each iteration in each video
        errors = []
        accs = []

        #Define iteration counter
        iter_counter = 0

        # Debugging print statements
        

        # DTEO 
        loop = True
        while loop == True:

            #Define a list of predictions for each video at the start of each iteration
            prediction_list = []

            txt_count = 0

            for txt in os.listdir(vid_path):

                # Only use 200 txts
                if txt_count < int(order):
                    txt_path = os.path.join(vid_path,txt)
                    
                    results = get_results(txt_path)
                    probs = get_probs(results)
                    prediction = get_count(probs,threshold)
                    #Add prediction to list of predictions
                    prediction_list.append(prediction)

                    txt_count += 1
            
            #increment iteration counter
            iter_counter += 1
            
            # Calculate mean of all predictions in vid
            mean_prediction = stat.mean(prediction_list)

            # Update threshold
            #altered_count = (1+0.1)*mean_prediction
            threshold = p(mean_prediction)

            # Add mean of the predictions to the count list
            counts.append(mean_prediction)

            # Calculate error for each iteration
            error = abs(mean_prediction-ground_truth)/ground_truth
            errors.append(error)

            if mean_prediction < ground_truth:
                acc = mean_prediction/ground_truth
            else:
                diff = mean_prediction-ground_truth
                acc = (ground_truth-diff)/ground_truth              
            accs.append(acc)

            
            # Stopping criteria
            # Max iterations = 8     
            #           
            if len(counts) > 1: # Ensure at least two iterations
                prop = counts[-2]/counts[-1] 
                
                # Max iterations

                if iter_counter > 12:
                    final_count = counts[-1] # Send last count
                    print("Max Iter")
                    loop = False

                # # Divergence
                #    Last         Second last
                elif counts[-1] <  counts[-2]:
                    final_count = counts[-2] # Send second last count
                    print("Diverged")
                    loop = False
                elif prop == 1.0:
                    final_count = counts[-1] # Send second last count
                    print("Converged")
                    loop = False

                # # Convergence        
                # if (counts[-1]/counts[-2])*100 < 80: # If last two counts are within 80% - break 
                #     final_count = counts[-2]
                #     break

                
        # Add final_count to counts_dict
        vid_error = abs(final_count-ground_truth)/ground_truth
        error_dict[vid] = vid_error
        all_vid_errors.append(vid_error)


        # if final_count < ground_truth:
        #     acc = final_count/ground_truth
        # else:
        #     diff = final_count-ground_truth
        #     acc = (ground_truth-diff)/ground_truth

        #count_dict_val[vid] = [final_count,ground_truth,error,errors,acc,counts]

    order_dict[order] =   error_dict  
    #print(len(all_vid_errors))
    av_vid_error = stat.mean(all_vid_errors)
    frame_error.append(av_vid_error)
```
